## Map Intent to Full Response Text

```python
# get full text of utter_ responses
intent_to_full_response = {} 

def map_intent_response_to_full_text(intents, reponsense):
    full = intents.copy()
    for key, values in full.items():
        new_vals = []
        for val in values:
            if 'action_' in val:
                continue
            else:
                new_vals.append(reponsense[val])
        full[key] = new_vals
        
    return full
        
intent_to_full_response = map_intent_response_to_full_text(full_response_intent,all_response_texts)
```

## API Calls

```python
def create_intent(project_id, display_name, training_phrases_parts, message_texts):
    """Create an intent of the given intent type."""
    
    from google.cloud import dialogflow

    intents_client = dialogflow.IntentsClient()

    parent = dialogflow.AgentsClient.agent_path(project_id)
    training_phrases = []
    
    for training_phrases_part in training_phrases_parts:
        print('phrase: '+training_phrases_part)
        all_parts = []
        
        if '@' not in training_phrases_part:
            all_parts = [dialogflow.Intent.TrainingPhrase.Part(text=training_phrases_part)]
            
        else:
            alias = None
            entity_type = None
            entity = False
            for s in training_phrases_part.split():
                
                if '@' in s:
                    alias = s[1:]
                    entity_type = s
                    entity = True
                    
                elif entity:
                    all_parts.append(dialogflow.Intent.TrainingPhrase.Part(text=(s+' '), entity_type = entity_type, 
                                                         alias = alias, user_defined=True))
                    entity = False
                else:
                    all_parts.append(dialogflow.Intent.TrainingPhrase.Part(text=s+' '))
                
        # Here we create a new training phrase for each provided part.
        training_phrase = dialogflow.Intent.TrainingPhrase(parts=all_parts)
        training_phrases.append(training_phrase)

    text = dialogflow.Intent.Message.Text(text=message_texts)
    message = dialogflow.Intent.Message(text=text)

    intent = dialogflow.Intent(
        display_name=display_name, training_phrases=training_phrases, messages=[message]
    )

    response = intents_client.create_intent(
        request={"parent": parent, "intent": intent}
    )

    print("Intent created: {}".format(response))
```

```python
# example intent creation
create_intent(PROJECT_ID, 'shirin', ['I want to go to a @area north restaurant', 
                                         'I want to go to a @area south sea'], intent_to_full_response['deny'])
```
